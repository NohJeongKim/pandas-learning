```python
from IPython.display import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df_copy=df.copy()
df_copy.head()
```

```python
df_copy["deck"]
```

```python
df_copy["deck"].mode() # 최빈값
```

```python
df_copy["deck"].mode()[0] # 최빈값을 지정하는 경우, 반드시 인덱스 지정해주기.
```

```python
df_copy["deck"].isnull().sum()
```

```python
df_copy["deck"].fillna(df_copy["deck"].mode()[0], inplace=True)
```

```python
df_copy["deck"].isnull().sum()
```
