```python
# 숫자에서 요일로 변경하는 함수 생성하기.
def num2str(x):
    if x==0:
        return "월"
    elif x==1:
        return "화"
    elif x==2:
        return "수"
    elif x==3:
        return "목"
    elif x==4:
        return "금"
    elif x==5:
        return "토"
    else:
        return "일"
```

```python
df_copy["요일"]=df_copy["date"].dt.dayofweek
df_copy.head()
```

```python
df_copy["요일"]=df_copy["요일"].apply(num2str)
df_copy
```
